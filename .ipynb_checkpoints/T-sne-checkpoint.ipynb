# Libraries

```python
import pandas as pd

from sklearn.preprocessing import scale
from sklearn.manifold import TSNE
import numpy as np

# Scaling
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
```

# Data Processing

Joining class list and feature list together

```python
df1 = pd.read_csv("Data/public-annotations.csv")
```

```python
df2 = pd.read_csv("Data/public_cohen_dataset_features.csv")
```

```python
df = df1.join(df2)
```

```python
print(df2.columns)
```

```python
df = df.drop(['S-A', 'S-B', 'S-C', 'S-D', 'S-E', 'S-F', 'J-A', 'J-B', 'J-C', 'J-D', 'J-E', 'J-F', 'J-Global'], axis = 1)
```

```python
df['s_class'] = 0
```

Assigning:  
Brixia 0-6: Class 1 | Less severe  
Brixia 7-12: Class 2 | Mild  
Brixia 13-18: Class 3 | Severe


```python
for i in range(192):
    if df.iloc[i, 1] <= 6:
        df.iloc[i, 109] = 1
        continue
    elif df.iloc[i, 1] <= 12:
        df.iloc[i, 109] = 2
        continue
    else:
        df.iloc[i, 109] = 3
```

```python
df
```

```python
df_nonames = df.drop('filename', axis = 1)
```

```python
df_nonames = df_nonames.drop('S-Global', axis = 1)
```

```python
df_nonames
```

Coverting to np array for TSNE function and scaling to be normalized

```python
#df_nonames = scale(df_nonames)
```

```python
scaler = MinMaxScaler()
df_nonames = pd.DataFrame(scaler.fit_transform(df_nonames), columns= df_nonames.columns, index = df_nonames.index)
df_nonames2 = scaler.fit_transform(df_nonames)
```

```python
df_nonames2 = df_nonames2.astype(np.float32)
```

```python
X = df_nonames2
```

```python
X
```

# Start T-SNE

```python
X_embedded = TSNE(n_components=2, perplexity = 10, learning_rate = .01, n_iter = 1000, metric = 'chebyshev').fit_transform(X)
```

```python
X_embedded.shape
```

```python
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c = df["s_class"])
```

```python
X_embedded = TSNE(n_components=3, init = 'random', n_iter = 3000).fit_transform(X)
```

```python
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], X_embedded[:, 2], c = df["s_class"])
```

```python
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(251, projection="3d")
ax.scatter(X_embedded[:, 0], X_embedded[:, 1], X_embedded[:, 2],c = df["s_class"])
```
